```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

csv_folder = "../output/cliphba_dynamic_v3/Things1854/emb"

n_timepoints = (800-100)//50 + 1
n_objects = 1854
n_dims = 66

emb_array = np.zeros((n_timepoints, n_objects, n_dims))
for timepoint in range(100, 850, 50):
    timepoint_index = (timepoint-100)//50
    emb_array[timepoint_index] = np.array(pd.read_csv(f"{csv_folder}/dynamic_embedding_{timepoint}ms.csv").iloc[:, 1:])

print(emb_array.shape)
```

```python
dim_array = np.mean(emb_array, axis=1)
print(dim_array.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming dim_array is a numpy array with shape (15, 66)
# Generate sample data for demonstration
# Remove the next line in your actual code

# Generate a figure and axis for the plot
fig, ax = plt.subplots()
timepoints = np.arange(dim_array.shape[0])

# Plot each line
for i in range(dim_array.shape[1]):
    ax.plot(timepoints, dim_array[:, i], label=f'Line {i+1}')

# Optionally, you can add labels, title, and legend
ax.set_xlabel('Timepoint')
ax.set_ylabel('Value')
ax.set_title('Line Plot of dim_array across 15 Timepoints')
# Uncomment the next line if you want to display a legend
# ax.legend()


plt.show()
```

```python
import numpy as np

def softmax(logits):
    exp_logits = np.exp(logits)
    return exp_logits / np.sum(exp_logits)

# Example usage
logits = np.array([1.0, 2.0, 3.0])
probabilities = softmax(logits)
print(probabilities)
```
